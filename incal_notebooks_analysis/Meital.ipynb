```python
import numpy as np
import pandas as pd
from collections import OrderedDict, Counter
import itertools
meital = pd.read_clipboard(parse_dates=['StartDate', 'StartTime'])
```

```python
grouped = meital.drop(columns='Cage').groupby(['StartDate', 'StartTime'])
group_all_mices = grouped.mean().reset_index()
date_time = pd.to_datetime(group_all_mices['StartDate'].astype('string') + " " + group_all_mices['StartTime'].astype('string'))
group_all_mices = group_all_mices.drop(columns=['StartDate', 'StartTime'])
group_all_mices = group_all_mices.set_index(date_time)
# group_all_mices[group_all_mices.index.hour >= 8].head(110)
```

```python
index = group_all_mices.index
```

```python
def replace_ids_to_group_id(pd_index):
    dates = sorted(set(pd_index.date))

    conditiones = [ logic 
        for date in dates 
        for logic in [
          (np.logical_and((pd_index.hour < 7), (pd_index.date == date))), 
          (np.logical_and((pd_index.hour >= 7), (pd_index.date == date)))] 
      ]
    choices = list(range(len(conditiones)))
    pd_index
    return np.select(conditiones, choices, pd_index.factorize()[0])
group_all_mices['days_exp'] = replace_ids_to_group_id(index)
```
